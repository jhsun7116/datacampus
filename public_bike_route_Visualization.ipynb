```
# 1. pandas 선언
import pandas as pd

# 2. Excel 파일 불러오기
df = pd.read_excel('/content/drive/My Drive/Colab Notebooks/0825_bike/자전거이동경로.xlsx')
df
```

지도에 표출한 사용자 정보 추출


```
df_route = df.loc[df['사용자 정보'] == 'User08']
df_route
```

PolyLine 지도 표시를 위한 좌표 list 만들기


```
# 지도에 표시하기 위해 folium 라이브러리 import 그리고 geo_df에 사용자 자전거 경로 정보 데이터프레임 할당
import folium

geo_df = df_route

# 경로를 표현하기 위해서 [ (위도1, 경도1), (위도2, 경도2), ...] 처럼 리스트안에 다중 튜플 구조를 만들기 위해서 테스트 코딩
lists = []
for n in geo_df.index:
    
    points = (geo_df.loc[n, '위도'], geo_df.loc[n, '경도'])
    lists.append(points)
print(lists)
```

경로 그리기

```
# 지도에 표시하기 위해 folium 라이브러리 import 그리고 geo_df에 사용자 자전거 경로 정보 데이터프레임 할당
import folium

map = folium.Map(location=[geo_df['위도'].mean(), geo_df['경도'].mean()], zoom_start=14)
folium.PolyLine(lists).add_to(map)

map.add_child(folium.LatLngPopup())

map
```

사용자 자전거 대여/반납 정보 추가

```
# 1. pandas 선언
import pandas as pd

# 2. Excel 파일 불러오기
user_df = pd.read_excel('사용자자전거대여반납.xlsx')

user_df.head()
```

한번에 경로 그리기 위한 전처리


```
Color = ['red', 'blue', 'green', 'purple', 'orange', 'darkred', 'darkblue', 'darkgreen', 'cadetblue', 'darkpurple', 'pink','gray', 'black']
users = ['User01','User02','User03','User04','User05','User06','User07','User08','User09','User10']
bike_path=[]

for user in users:
  temp = df.loc[df['사용자 정보'] == user]
  temp_list=[]
  for i in temp.index:
    latlng=[]
    latlng.append(temp['위도'][i])
    latlng.append(temp['경도'][i])
    temp_list.append(latlng)

  bike_path.append(temp_list)
  
#print(bike_path)
```

지도에 사용자 정보 표출

```
map = folium.Map(location=[df['위도'].mean(), df['경도'].mean()], zoom_start=11)


for k,each_path in enumerate(bike_path):
  folium.PolyLine(locations = each_path, tooltip = 'PolyLine',color=Color[k]).add_to(map) 

for n in user_df.index[:10]:
  folium.Marker([user_df.loc[n, '대여위도'], user_df.loc[n, '대여경도']],popup=user_df.loc[n,'사용자 정보'], 
                icon=folium.Icon(icon="fa-play-circle",color=Color[n],prefix='fa')).add_to(map)
  folium.Marker([user_df.loc[n, '반납위도'], user_df.loc[n, '반납경도']],popup=user_df.loc[n,'사용자 정보'], 
                icon=folium.Icon(icon="fa-stop",color=Color[n],prefix='fa')).add_to(map)

map
```
